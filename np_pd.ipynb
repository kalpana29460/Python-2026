```python
import numpy as np
import pandas as pd
data = {
    "Name":["Ravi","kamala","kalpana","suja","vimala","Ravi","Kiran","Anu"],
    "Age":[25,30,22,np.nan,29,25,5,120],
    "Salary":[50000,np.nan,45000,60000,52000,50000,150000,30000],
    "Department":[2,5,1,4,3,2,20,1]
}

df=pd.DataFrame(data)

print("Original Data:\n")
print(df)
```

```python
df.drop_duplicates(inplace=True)
print(df)
```

```python
df = df[df["Age"] < 100]
print(df)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df["Department"] = le.fit_transform(df["Department"])
print(df)
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Salary"] = df["Salary"].fillna(df["Salary"].median())

print(df)
```

```python


from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df[["Age", "Salary",]] = scaler.fit_transform(
    df[["Age", "Salary", ]]
)
print(df)
```

```python


print("\n🟢 Cleaned Data:\n")
print(df)
```

```python
import matplotlib.pyplot as plt
plt.hist(df["Salary"], bins=5)
plt.title("Salary Distribution After Cleaning")
plt.show()
```
