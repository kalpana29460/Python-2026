```python


sales = 0
orders = 0
feedbacks = 0

print("Dashboard Categories:")
print("1. Sales")
print("2. Orders")
print("3. Feedbacks")

choice = int(input("Enter category number to increment count: "))

if choice == 1:
    sales += 1
elif choice == 2:
    orders += 1
elif choice == 3:
    feedbacks += 1
else:
    print("Invalid choice")

print("\n--- DASHBOARD ---")
print("Sales    :", sales)
print("Orders   :", orders)
print("Feedbacks:", feedbacks)
```
